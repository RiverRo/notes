- Note: python and numpy random modules are pseudo-random
	- meaning with a seed the random events are reproducible

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Random
- Individual values/events

```python
# 1 = heads. 0 = tails.
rolls = []  
for i in range(100):
  roll = np.random.randint(0,2)
  rolls.append(1 if roll == 1 else 0)  
plt.plot(rolls)
plt.show()
```

## Random Walk
- Next value takes into accout previous value

```python
rolls_walk = [0]
for s in range(100):
  roll = np.random.randint(0,2)
  step = rolls_walk[-1] # initial to previous step
  if roll == 1:
    step = step + 1
  rolls_walk.append(step)
plt.plot(rolls_walk)
plt.show()
```
